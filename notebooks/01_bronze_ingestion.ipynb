```python
from pyspark.sql import SparkSession

orders = [
    (1001, 101, 501, 500.0, "2026-06-23 10:00:00"),
    (1002, 102, 502, 700.0, "2026-06-23 10:01:00"),
    (1001, 101, 501, 500.0, "2026-06-23 10:00:00"),  # duplicate
    (1003, None, 503, 900.0, "2026-06-23 10:02:00")  # bad record
]

columns = [
    "order_id",
    "customer_id",
    "product_id",
    "amount",
    "event_time"
]

df = spark.createDataFrame(orders, columns)

display(df)
```

```python
spark.sql("""
          create database if not exists ecommerce_project
""")
```

```python
df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("ecommerce_project.bronze_orders")
```

```python
display(
    spark.sql("""
              select * from ecommerce_project.bronze_orders""")
)
```
